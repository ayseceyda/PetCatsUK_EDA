# **Giriş**

Kediler, yaşam koşulları ülkeden ülkeye değişiklik göstermekle birlikte, evlerimizde ya da sokaklarda gündelik yaşamın bir parçası olarak varlıklarını sürdüren sevimli dostlarımızdır. Evcil kedilerin ağırlıklı yaşam alanları ise evler olmasına karşın; farklı iç ve dış etkenlerle birlikte tetiklenebilecek sosyalleşme ihtiyaçları, onları dış dünyaya karışmaya yönlendirebilmektedir. Bu yazıda, bu yönlendirici etkilerin neler olduğu ve dış dünyaya karışma dürtüsü ile aralarındaki ilişki incelenmiştir.

İnceleme sırasında kullanılan veri setleri, 2020 tarihli makaleden kökenlenmiştir. Veri setleri içerisinde, 2013–2017 yılları arasında 4 farklı ülkedeki (Amerika, Birleşik Krallık, Avustralya ve Yeni Zelanda) evcil kedilere, GPS sensörü takılması ile toplanmış birçok tipte veri bulunmaktadır. Keşifsel veri analizi için kullanılacak olan kaynakta ise sadece Birleşik Krallık (UK) verisi mevcuttur.

Tüm kediler, veri seti gözünden kedi cinsleri noktasında ayrışmayıp; birbirlerinden konum, zaman, hız, üreme yetisi, avlanma, cinsiyet, yaş ya da farklı mama çeşitlerinin tüketimi noktasında ayrışmaktadır.
Veri analizi sırasında kullanılacak olan kaynaklar, 2 farklı veri setinden oluşmaktadır:

1. **cats_uk.csv:** Kedilere takılan sensörler üzerinden elde edilen zaman, konum ve hız bilgilerini içerir.
2. **cats_uk_reference.csv:** Her bir kediye ait yaş, cinsiyet, kısırlaştırma, avlanma, beslenme gibi özelleştirilmiş bilgileri içerir.

**Hipotez**

  Bu yazıda makalenin ışığında üretilen hipotez, keşifsel veri analizi kullanılarak doğru ya da yanlışlığı noktasında kanıtlanmak istenmiştir. Sonuçlar ve tartışma kısmında çıktıları paylaşılan hipotezin 3 maddesi ise şu şekildedir:

  * " Evcil kedilerin az ya da fazla mesafe katetmesi, Avlanma ve Üreme Yetisi gibi karakteristiklerle olumlu bir ilişki içerisindedir. "

  Dönemsel değişen hormon seviyelerine bağlı olarak kedilerde dışarıda daha fazla zaman geçirme ve mesafe katetmeye yönelik bir içgüdü oluşabilir. Buradaki dönem değişkeni, verideki 6 ay kısıtından dolayı araştırmada sadece 2 sezon ile (yaz, sonbahar) sınırlı kalmıştır.

  Araştırmada yer verilmeyen kedi cinsi bilgisine bağlı olarak avlanma güdü ve alışkanlığı, dışarıda vakit geçirmeye teşvik edici bir etken olabilir. Üreme iç güdüsüne benzer şekilde, avlanma ile de aralarında dönemsel etki oluşması beklenebilir.

* " Evcil kedilerin (ev harici izlenen) hareket iz düşümleri, kıyı kesimlere doğru yoğunlaşır. "

  Dışarı çıkan evcil kedilerdeki hareket yoğunluğunun, haritadaki kıyı kesimlere doğru arttığı gözlemlenmek istenmektedir. Bu yönelimde, olası kuş ya da balık avı gibi avcılık faaliyetlerinin etkili olabileceği düşünülmektedir.

```python
# 1 CATS UK REFERENCE DATASETİ İNCELEMESİ

# Kütüphanelerin yüklenmesi
import pandas as pd

cats_uk_ref_url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2023/2023-01-31/cats_uk_reference.csv"
cats_uk_ref = pd.read_csv(cats_uk_ref_url)

# Unique değerleri görüntüleme
# for col in cats_uk_ref:
#   print(cats_uk_ref[col].unique())

# İşlemler öncesinde veri boyutunu görüntüleme
# rows = len(cats_uk_ref.axes[0])
# cols = len(cats_uk_ref.axes[1])
# print("Number of Rows: ", rows) # Number of rows: 101
# print("Number of Columns: ", cols) # Number of columns: 16

# Null değerleri görüntüleme
# print(cats_uk_ref.isna().sum())

# Bazı kolonlarda null değerler mevcuttur:
# hunt                              9
# animal_reproductive_condition     3
# food_other                       10
# age_years                         1

# hunt (avlanma) bilgisi null gelen kediler
cats_uk_ref = cats_uk_ref.fillna(-1)
cats_uk_ref_hunt_null = cats_uk_ref[cats_uk_ref["hunt"] == -1 ]
#print(cats_uk_ref_hunt_null['tag_id'])

# animal_reproductive_condition (kısırlaştırma bilgisi) null gelen kediler
cats_uk_ref = cats_uk_ref.fillna(-1)
cats_uk_ref_animal_reproductive_condition_null = cats_uk_ref[cats_uk_ref["animal_reproductive_condition"] == -1 ]
#print(cats_uk_ref_animal_reproductive_condition_null['tag_id'])

# food_other (kuru ya da yaş mama dışındaki besinle beslenme bilgisi) null gelen kediler
cats_uk_ref = cats_uk_ref.fillna(-1)
cats_uk_ref_food_other_null = cats_uk_ref[cats_uk_ref["food_other"] == -1 ]
#print(cats_uk_ref_food_other_null['tag_id'])

# age (yaş) bilgisi null gelen kediler
cats_uk_ref = cats_uk_ref.fillna(-1)
cats_uk_ref_age_years_null = cats_uk_ref[cats_uk_ref["age_years"] == -1 ]
#print(cats_uk_ref_age_years_null['tag_id'])

# Dataframe' den çıkartılacak toplam 23 kedi ismi:
cats_to_remove = ["Ares",
"Wilfred-Tag",
"Lightening Bugg-Tag",
"Jezebel-Tag",
"Tipsy-Tag",
"Maggie-Tag",
"Lily-Tag",
"Pants-Tag",
"Neil-Tag",
"Meg-Tag",
"Amber-Tag",
"Charlie3-Tag",
"Morpheus-Tag",
"Nettle-Tag",
"Fonzie-Tag",
"Barney-Tag",
"Rusty-Tag",
"Bear-Tag",
"Chloe-Tag",
"Jago2-Tag",
"Roger-Tag",
"Gracie_2-Tag",
"Johnny-Tag"]

cats_uk_ref = cats_uk_ref[~cats_uk_ref['tag_id'].isin(cats_to_remove)]

# age (yaş) bilgisi kolonunun veri tipi int' e çevrilmiştir.
cats_uk_ref['age_years'] = cats_uk_ref['age_years'].astype(int)

# animal_taxon kolonu sadece 'Felis catus' değerini içerdiği ve analizlerde kullanılmayacağı için drop edilmiştir.
cats_uk_ref = cats_uk_ref.drop("animal_taxon", axis='columns')

# study_site kolonu sadece 'UK' değerini içerdiği ve analizlerde kullanılmayacağı için drop edilmiştir.
cats_uk_ref = cats_uk_ref.drop("study_site", axis='columns')

# timestamp kolonundaki ISO 8601 formatı datetime' a çevrilmiştir
cats_uk_ref['deploy_on_date'] = pd.to_datetime(cats_uk_ref['deploy_on_date'], errors='coerce')
cats_uk_ref['deploy_off_date'] = pd.to_datetime(cats_uk_ref['deploy_off_date'], errors='coerce')
# YYYY-MM-DD HH:MM:SS
cats_uk_ref['deploy_on_date'] = cats_uk_ref['deploy_on_date'].dt.strftime('%Y-%m-%d %H:%M:%S')
cats_uk_ref['deploy_off_date'] = cats_uk_ref['deploy_off_date'].dt.strftime('%Y-%m-%d %H:%M:%S')

# Elemeler sonrası unique değer kontrolü yapılmıştır.

for col in cats_uk_ref:
    print(cats_uk_ref[col].unique())

# Elemeler sonrası veri boyutu kontrolü yapılmıştır.

rows = len(cats_uk_ref.axes[0])
cols = len(cats_uk_ref.axes[1])

print("Final Number of Rows: ", rows) # Number of Rows:  101 - 23 = 78
print("Final Number of Columns: ", cols) # Number of Columns:  14

# Elemeler sonrası null değerler görüntülenmiştir.
print(cats_uk_ref.isna().sum())

# Veri tiplerini görüntüleme
print(cats_uk_ref.dtypes)
```

```python
# 2 CATS UK DATASETİ İNCELEMESİ

# Kütüphanelerin yüklenmesi
import pandas as pd

# Cats UK verisetinin yüklenmesi
cats_uk_url = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2023/2023-01-31/cats_uk.csv"
cats_uk = pd.read_csv(cats_uk_url)

# Unique değerleri görüntüleme
# for col in cats_uk:
#    print(cats_uk[col].unique())

# Null değerleri görüntüleme
# print(cats_uk.isna().sum()) # Null değer yoktur

# İşlemler öncesinde veri boyutunu görüntüleme
# rows = len(cats_uk.axes[0])
# cols = len(cats_uk.axes[1])
# print("Number of Rows: ", rows) # Number of rows: 18.215
# print("Number of Columns: ", cols) # Number of columns: 11

# Reference verisetindeki 23 null değer taşıyan kedi ismi, bu verisetinden de çıkarılmıştır.
cats_uk = cats_uk[~cats_uk['tag_id'].isin(cats_to_remove)]

# study_name kolonu sadece tek bir değer (Pet Cats United Kingdom) içerdiği ve analizlerde kullanılmayacağı için drop edilmiştir.
cats_uk = cats_uk.drop("study_name", axis='columns')

# Outlier TRUE değerindeki satırlar elenmiş ve ardından outlier bilgisini içeren kolonlar drop edilmiştir.

cats_uk = cats_uk[cats_uk['algorithm_marked_outlier'] != True]
cats_uk = cats_uk[cats_uk['manually_marked_outlier'] != True]

cats_uk = cats_uk.drop("algorithm_marked_outlier", axis='columns')
cats_uk = cats_uk.drop("manually_marked_outlier", axis='columns')

# visible kolonu algorithm_marked_outlier ya da manually_marked_outlier değerlerinden en az birisi TRUE (outlier) olduğunda FALSE ; değerlerden her ikisi de FALSE (outlier olmadığı) durumda ise TRUE' dur.
# Outlier kolonlarından gerekli elemeler yapıldığı için bu kolonun kullanımına gerek kalmamıştır.

cats_uk = cats_uk.drop("visible", axis='columns')

# event_id kolonu her bir köpek için unique' tir. Analize katkı sunmayacağı için drop edilmiştir.

cats_uk = cats_uk.drop("event_id", axis='columns')

# timestamp kolonundaki ISO 8601 formatı datetime' a çevrilmiştir: YYYY-MM-DD HH:MM:SS
cats_uk['timestamp'] = pd.to_datetime(cats_uk['timestamp'], errors='coerce')
# YYYY-MM-DD HH:MM:SS
cats_uk['timestamp'] = cats_uk['timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S')

# Elemeler sonrası unique değer kontrolü yapılmıştır.

for col in cats_uk:
    print(cats_uk[col].unique())

# Elemeler sonrası veri boyutu kontrolü yapılmıştır.

rows = len(cats_uk.axes[0])
cols = len(cats_uk.axes[1])

print("Final Number of Rows: ", rows) # Number of Rows:  13.681 (23 null değer taşıyan kedi cinslerine ait çoklayan tüm veriler silinmiştir.)
print("Final Number of Columns: ", cols) # Number of Columns:  6

# Elemeler sonrası null değer kontrolü yapılmıştır.
print(cats_uk.isna().sum()) # Null değer yoktur

# Veri tiplerini görüntüleme
print(cats_uk.dtypes)
```

```python
# 3 VERİSETLERİNİN BİRLEŞTİRİLMESİ

# tag_id ortak kolonunu taşıyan "CATS UK REFERENCE" ve "CATS UK" verisetleri inner join yardımıyla birleştirilmiştir.
final_dataset = pd.merge(cats_uk, cats_uk_ref, on='tag_id', how='inner')

# Birleşim sonrası tag_id kolonu drop edilerek yerine daha anlamlı olan animal_id kolonu kullanılmıştır.
final_dataset = final_dataset.drop("tag_id", axis='columns')

# Kedi Yaşları:

#Kitten (0-6 months) NOK
#Junior (7 months-2 years) NOK
#Adult (3-6 years) OK
#Mature (7-10 years) OK

# Hız ve konum ölçümleneceği için, yavru ya da erken bebeklik döneminde olan kediler verisetinden elenmiştir.
final_dataset = final_dataset.loc[final_dataset['age_years'] >= 3]

# Birleşim sonrası null kontrolü yapılmıştır.
print(final_dataset.isna().sum())

# Veri Boyutu
rows = len(final_dataset.axes[0])
cols = len(final_dataset.axes[1])

print("Final Number of Rows: ", rows)
print("Final Number of Columns: ", cols)

# Veri tiplerini görüntüleme
print(final_dataset.dtypes)

final_dataset
#final_dataset.to_csv("final_dataset.csv", index=False)
```

```python
# 2. TIMESERIES GRAPH

# Kütüphanelerin yüklenmesi
import pandas as pd
import matplotlib.pyplot as plt
import mpld3
from mpld3 import plugins
from IPython.core.display import display, HTML

# 'yearmonth' kolonunun oluşturulması
final_dataset['timestamp'] = pd.to_datetime(final_dataset['timestamp'])
final_dataset['yearmonth'] = final_dataset['timestamp'].dt.to_period('M')

# her yıl-ay için ortalama hız değerinin hesaplanması
animal_avg_speed = final_dataset.groupby(['animal_id', 'yearmonth'])['ground_speed'].mean().reset_index()

# her yıl-ay için en hızlı ve en yavaş kedilerin belirlenmesi
fastest_per_month = animal_avg_speed.loc[animal_avg_speed.groupby('yearmonth')['ground_speed'].idxmax()]
slowest_per_month = animal_avg_speed.loc[animal_avg_speed.groupby('yearmonth')['ground_speed'].idxmin()]

# dinamik tooltips detaylarının oluşturulması
fastest_per_month = pd.merge(
    fastest_per_month,
    final_dataset[['animal_id', 'animal_sex', 'hunt', 'animal_reproductive_condition', 'age_years']],
    on='animal_id',
    how='left'
)
slowest_per_month = pd.merge(
    slowest_per_month,
    final_dataset[['animal_id', 'animal_sex', 'hunt', 'animal_reproductive_condition', 'age_years']],
    on='animal_id',
    how='left'
)

fig, ax = plt.subplots(figsize=(10, 5))

# en hızlı kediler
fastest_line, = ax.plot(
    fastest_per_month['yearmonth'].dt.to_timestamp(),
    fastest_per_month['ground_speed'],
    marker='o',
    label='Fastest',
    color='#1f78b4',
    markersize=10
)

# en hızlı kediler için dinamik etiketleme
fastest_labels = [
    f"<b>Name:</b> {row['animal_id']}<br>"
    f"<b>Sex:</b> {row['animal_sex']}<br>"
    f"<b>Age:</b> {row['age_years']} years<br>"
    f"<b>Breeding:</b> {row['animal_reproductive_condition']}<br>"
    f"<b>Hunt:</b> {row['hunt']}"
    for _, row in fastest_per_month.iterrows()
]
fastest_tooltips = plugins.PointHTMLTooltip(fastest_line, labels=fastest_labels, css="font-size: 12px; color: #333; background: #f0f0f0; padding: 5px; border-radius: 5px;")
mpld3.plugins.connect(fig, fastest_tooltips)

# en yavaş kediler
slowest_line, = ax.plot(
    slowest_per_month['yearmonth'].dt.to_timestamp(),
    slowest_per_month['ground_speed'],
    marker='o',
    label='Slowest',
    color='#FFA500',
    markersize=10
)

# en yavaş kediler için dinamik etiketleme
slowest_labels = [
    f"<b>Name:</b> {row['animal_id']}<br>"
    f"<b>Sex:</b> {row['animal_sex']}<br>"
    f"<b>Age:</b> {row['age_years']} years<br>"
    f"<b>Breeding:</b> {row['animal_reproductive_condition']}<br>"
    f"<b>Hunt:</b> {row['hunt']}"
    for _, row in slowest_per_month.iterrows()
]
slowest_tooltips = plugins.PointHTMLTooltip(slowest_line, labels=slowest_labels, css="font-size: 12px; color: #333; background: #fffbea; padding: 5px; border-radius: 5px;")
mpld3.plugins.connect(fig, slowest_tooltips)

# Görünüm ayarlamaları
tooltip_css = """
font-size: 25px;
color: #333;
background: #f0f0f0;
padding: 5px;
border-radius: 5px;
border: 1px solid #ccc;
box-shadow: 2px 2px 5px rgba(0, 0, 0, 0.1);
"""

# en hızlı kediler için dinamik etiketleme
fastest_tooltips = plugins.PointHTMLTooltip(
    fastest_line,
    labels=fastest_labels,
    css=tooltip_css
)
mpld3.plugins.connect(fig, fastest_tooltips)

# en yavaş kediler için dinamik etiketleme
slowest_tooltips = plugins.PointHTMLTooltip(
    slowest_line,
    labels=slowest_labels,
    css=tooltip_css
)
mpld3.plugins.connect(fig, slowest_tooltips)

# Görünüm ayarlamaları
font1 = {'family': 'sans-serif', 'size': 15, 'color': 'grey'}

plt.grid(color='grey', linestyle='-', linewidth=0.5, alpha=0.2)

ax.set_title('Average Ground Speed for Cats Over Time', fontsize=17, pad=25)
ax.set_xlabel('Months of 2017', fontdict=font1, labelpad=10)
ax.set_ylabel('Average Ground Speed (m/s)', fontdict=font1, labelpad=10)

# Legend' ın Hizalanması
ax.legend(
    loc='center left',
    bbox_to_anchor=(1, 0.5),
    fontsize=10,
    edgecolor="white"
)
ax.grid(True)

# İnteraktif grafiğin HTML formatına dönüştürülmesi
mpld3_html = mpld3.fig_to_html(fig)

# HTML formatında notebook'a gömülmesi ve ortalanması
display(HTML(f"""
<div style="display: flex; justify-content: center; align-items: center; height: 100%; margin-top: 20px;">
    {mpld3_html}
</div>
"""))

# Grafiğin Gösterimi
plt.close(fig)
```

```python
# VERİ HAZIRLIĞI: # Her bir kedi için kümülatif mesafenin hesaplanması (EN ÇOK 6)

import pandas as pd
from geopy.distance import geodesic

final_dataset['coords'] = list(zip(final_dataset['location_lat'], final_dataset['location_long']))

data = final_dataset.sort_values(['animal_id', 'timestamp'])

# Her bir kedi için toplam mesafenin hesaplanması
def calculate_total_distance(group):
    distances = [
        geodesic(group.iloc[i], group.iloc[i + 1]).meters
        for i in range(len(group) - 1)
    ]
    return sum(distances)

total_distances = data.groupby('animal_id')['coords'].apply(calculate_total_distance).reset_index(name='total_distance')

# En Fazla Mesafe Kateden Kedilerin Listelenmesi
top_animals = total_distances.nlargest(6, 'total_distance')

print(top_animals)
```

```python
# VERİ HAZIRLIĞI: # Her bir kedi için kümülatif mesafenin hesaplanması (EN AZ 6)

import pandas as pd
from geopy.distance import geodesic

data['coords'] = list(zip(data['location_lat'], data['location_long']))

data = data.sort_values(['animal_id', 'timestamp'])

# Her bir kedi için toplam mesafenin hesaplanması
def calculate_total_distance(group):
    distances = [
        geodesic(group.iloc[i], group.iloc[i + 1]).meters
        for i in range(len(group) - 1)
    ]
    return sum(distances)

total_distances = data.groupby('animal_id')['coords'].apply(calculate_total_distance).reset_index(name='total_distance')

# En Az Mesafe Kateden Kedilerin Listelenmesi
bottom_animals = total_distances.nsmallest(6, 'total_distance')

print(bottom_animals)
```

```python
# 3. GEOGRAPHICAL + ANIMATED GRAPH

# Kütüphanelerin yüklenmesi
import folium
from folium import Html, IFrame, Element
from folium.plugins import FloatImage
import pandas as pd
from geopy.distance import geodesic

# Seçili kedilerin eklenmesi
selected_cats = {
    "Ginge": "#4898c5",
    "Teddy": "#FFA500",
    "Tom": "#069f6f",
    "Max": "#a46d2d",
    "Maxwell": "#c3a46d",
    "Merlin": "#6fab9f",

    "Magic": "#4898c5",
    "Ebby": "#FFA500",
    "Tigger": "#069f6f",
    "Flash": "#a46d2d",
    "Charlie2": "#c3a46d",
    "Henry": "#6fab9f"
}

# Haritanın Ortalanması
mean_lat = final_dataset['location_lat'].mean()
mean_long = final_dataset['location_long'].mean()

# Harita oluşturma (kare oranında boyut ayarı)
m = folium.Map(
    location=[mean_lat, mean_long],
    zoom_start=11,
    width="55%",
    height="60%"
)

# Seçili kedilerin aldığı yolun harita üzerinde işaretlenmesi
for cat, color in selected_cats.items():
    cat_data = final_dataset[final_dataset['animal_id'] == cat]
    if len(cat_data) > 1:

        # Yolun gösterimi
        locations = list(zip(cat_data['location_lat'], cat_data['location_long']))

        # Yolculuğun toplam mesafesinin hesaplanması
        total_distance_km = sum(
            geodesic(locations[i], locations[i + 1]).km for i in range(len(locations) - 1)
        )

        # Ek bilgilerin 'hover' sonrası gösterimi
        tooltip_text = (
            f"<b>Name:</b> {cat}<br>"
            f"<b>Sex:</b> {cat_data['animal_sex'].iloc[0]}<br>"
            f"<b>Hunt:</b> {cat_data['hunt'].iloc[0]}<br>"
            f"<b># of Prey:</b> {cat_data['prey_p_month'].iloc[0]}<br>"
            f"<b>Breeding:</b> {cat_data['animal_reproductive_condition'].iloc[0]}<br>"
            f"<b>Total Path:</b> {total_distance_km:.2f} km"
        )

        polyline = folium.PolyLine(
            locations=locations,
            color=color,
            weight=3,
            tooltip=folium.Tooltip(tooltip_text)
        )
        polyline.add_to(m)

        # Yolculuğun başlangıç noktası
        start_location = locations[0]
        folium.Marker(
            location=start_location,
            icon=folium.Icon(color="green", icon="play", prefix="fa"),
            tooltip=f"{cat} (Start)"
        ).add_to(m)

        # Yolculuğun bitiş noktası
        end_location = locations[-1]
        folium.Marker(
            location=end_location,
            icon=folium.Icon(color="red", icon="stop", prefix="fa"),
            tooltip=f"{cat} (End)"
        ).add_to(m)

# Grafiğe ait veri İşaretleyicileri ve Görsel Kanallar
text_content = """
<div style="position: absolute; left: 60%; top: 10%; width: 30%; height: auto; background-color: #f9f9f9; border: 3px solid green; padding: 10px; font-size: px;">
    <p><b>Şekil 3 : </b> En az ve en çok mesafe kateden 12 evcil kediye ait veriler, coğrafi ve animasyonlu grafikler birleştirilerek detaylandırılmıştır. Enlem ve boylam bilgisi kullanılarak ilgili
    zaman aralıkları içerisindeki toplam katedilen km bilgisi, imleç ile üzerine gelindiğinde ortaya çıkacak şekilde etiketlenmiştir. </p>
    <p><b>Veri işareti:</b> Alan, Çizgi </p>
    <p><b>Kullanılan görsel kanallar:</b></p>
    <ul>
        <li><b>Şekil 3 X-pozisyonu:</b> Enlem bilgisinin harita üzerinde gösterimi</li>
        <li><b>Şekil 3 Y-pozisyonu:</b> Boylam bilgisinin harita üzerinde gösterimi</li>
    </ul>
    <p><b>Renk:</b> Seçili her bir evcil kedi için ziyaret edilen güzergahın işaretlenmesi</p>
    <p><b>Büyüklük:</b> Seçili her bir evcil kedi için ziyaret edilen güzergahın toplam alanı </p>
    <p><b>Veri Etiketi 1:</b> Seçili her bir evcil kedi için ziyarete başlangıç ve bitiş noktalarının işaretlenmesi </p>
    <p><b>Veri Etiketi 2:</b> Seçili her bir evcil kedi için isim, cinsiyet, ortalama hız, avlanma bilgisi, üreme yetisi bilgisi ve toplam katedilen mesafenin gösterimi (km) </p>
    <p><b>Pozisyon:</b> Seçili her bir evcil kedi için  ziyaret edilen enlem ve boylam koordinatlarına bakılarak harita üzerinde işaretleme yapılması </p>

</div>
"""

folium.map.LayerControl().add_to(m)
m.get_root().html.add_child(folium.Element(text_content))

# Haritayı göster
m
```
